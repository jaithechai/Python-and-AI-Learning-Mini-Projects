# Anomaly Detection in Data Visualization Using Python

This Colab notebook provides an overview of anomaly detection techniques using data visualization with Python. It demonstrates how to analyze, detect, and visualize outliers in datasets by calculating statistical parameters and using visual tools like histograms and line plots.

## Key Concepts Covered

### Data Analysis and Anomaly Detection
- **Mean and Standard Deviation Calculation**: Basic statistical measures to understand data distribution.
- **Anomaly Identification**: Detecting data points that are more than 3 standard deviations from the mean, indicating outliers.

### Data Visualization
- **Histograms**: Display the distribution of data, helping to identify unusual values.
- **Gaussian Distribution Curve**: Overlayed on histograms to illustrate the normal distribution of data.
- **Line Plots**: Visualize trends over time, marking anomalies for easier detection.
- **Scatter Plots**: Highlight outliers to show which data points deviate from the expected range.

### Code Explanation and Visual Interpretation
- **Step-by-Step Coding**: Provides a detailed breakdown of how to load, process, and visualize data.
- **Customizable Parameters**: Adjust parameters such as thresholds for anomaly detection.
- **Easy-to-Understand Plots**: Visual aids to quickly identify patterns and anomalies in the data.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Temperature Example
# Sample data
temperatures = [68, 70, 69, 71, 72, 73, 69, 68, 67, 71, 74, 69, 70, 71, 68, 69, 67, 66, 72, 71, 69, 70, 68, 67, 66, 65, 64, 63, 62, 75]  # Last value is an anomaly

# Calculate mean and standard deviation
mean_temp = np.mean(temperatures)
std_temp = np.std(temperatures)

# Identify anomalies
anomalies = [temp for temp in temperatures if temp < mean_temp - 3 * std_temp or temp > mean_temp + 3 * std_temp]

# Visualize
x = np.linspace(mean_temp - 4*std_temp, mean_temp + 4*std_temp, 1000)
y = (1/(np.sqrt(2 * np.pi * std_temp**2))) * np.exp(-(x - mean_temp)**2 / (2 * std_temp**2))

plt.hist(temperatures, bins=10, alpha=0.6, color='g', edgecolor='black', density=True)
plt.plot(x, y, color='r')
plt.xlabel('Temperature (Â°F)')
plt.ylabel('Frequency')
plt.title('Temperature Distribution')
plt.scatter(anomalies, [0]*len(anomalies), color='red')
plt.show()
```

```python
# Sample website traffic dataset (number of visits per day)
# Sample data
network_traffic = [100, 102, 98, 101, 99, 103, 100, 101, 102, 100, 101, 99, 100, 98, 97, 102, 100, 101, 99, 98, 97,150,150]  # 150 is an anomaly
# Calculate mean and standard deviation
mean_traffic = np.mean(network_traffic)
std_traffic = np.std(network_traffic)
# Identify anomalies
anomalies = [traffic for traffic in network_traffic if traffic < mean_traffic - 3 * std_traffic or traffic > mean_traffic + 3 * std_traffic]

# Visualize
time = range(len(network_traffic))
plt.plot(time, network_traffic, marker='o', linestyle='-', color='b')
plt.axhline(mean_traffic, color='r', linestyle='dashed', linewidth=1)
plt.axhline(mean_traffic + 3*std_traffic, color='g', linestyle='dashed', linewidth=1)
plt.axhline(mean_traffic - 3*std_traffic, color='g', linestyle='dashed', linewidth=1)
plt.scatter([i for i, traffic in enumerate(network_traffic) if traffic in anomalies], anomalies, color='red')
plt.xlabel('Time (minutes)')
plt.ylabel('Network Traffic (requests/second)')
plt.title('Network Traffic Over Time')
plt.show()
```

```python
steps=np.array([10200, 9800, 11000, 10100, 10500, 11500, 8000, 9500, 10700, 15000,])
steps_mean=np.mean(steps)
steps_std=np.std(steps)
anomalies=[i for i in steps if i>steps_mean + 3 * steps_std or i < steps_mean - 3 * steps_std]

x=np.linspace(steps_mean - 3 * steps_std, steps_mean + 3 * steps_std, 1000)
y=(1/(np.sqrt(2 * np.pi * steps_std**2))) * np.exp(-(x - steps_mean)**2 / (2 * steps_std**2))

plt.hist(steps, bins=10, alpha=0.6, color='g', edgecolor='black', density=True)
plt.plot(x,y, color='r')
plt.xlabel('Steps')
plt.ylabel('Frequency')
plt.title('Steps Distribution')
plt.scatter((anomalies), [0]*len(anomalies), color='red')
plt.show()

time = range(len(steps))
plt.plot(time, steps, marker='o', linestyle='-', color='b')
plt.axhline(steps_mean, color='r', linestyle='dashed', linewidth=1)
plt.axhline(steps_mean + 3*steps_std, color='g', linestyle='dashed', linewidth=1)
plt.axhline(steps_mean - 3*steps_std, color='g', linestyle='dashed', linewidth=1)
plt.scatter([i for i, traffic in enumerate(steps) if traffic in anomalies], anomalies, color='red')
plt.xlabel('Time (minutes)')
plt.ylabel('Steps /second)')
plt.title('Steps Over Time')
plt.show()
```

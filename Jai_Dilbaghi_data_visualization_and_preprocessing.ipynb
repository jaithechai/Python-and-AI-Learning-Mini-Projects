```python
#pip install ucimlrepo
!pip install ucimlrepo
import ucimlrepo
chK=ucimlrepo.fetch_ucirepo(id=336)
x=chK.data.features
y=chK.data.target
dt=pd.concat([x,y],axis=1)
dt.info()
```

```python
cat_features=dt.dtypes==object
cat_cols=dt.columns[cat_features].tolist()
cat_cols
dt[cat_cols].dropna()
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dt[cat_cols]=dt[cat_cols].apply(le.fit_transform)
dt[cat_cols].head()
```
